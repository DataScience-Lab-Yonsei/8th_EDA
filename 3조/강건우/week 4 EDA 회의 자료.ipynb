```python
def numeric_process(data):
  
   
    data['BIRTH_MINUS_EMPLOYED'] = data['DAYS_BIRTH'] - data['DAYS_EMPLOYED']
    
    return data 
```

```python
Train24 = numeric_process(Train21)
Train24
```

```python
Train25=Train24.sort_values(ascending=False,by='DAYS_EMPLOYED').head(98)
Train25
print(Train25.mean())
```

```python
Train26=Train24.sort_values(ascending=False,by='DAYS_EMPLOYED').tail(98)
Train26
print(Train26.mean())
```

```python
#오히려 일을 늦게 시작한 공무원들의 평균 신용도가 비교적 양호하다. 평균 소득에도 큰 차이가 없다 (다만 오래 공무원직을 수행 사람이 약간 높다)
#다른일에 종사하다가 고위 공무직을 하게 되는 경우 신용도가 높을 수 있다.

#공무원직에 오래 머무룰수록 평균 소득은 높아지는 경향이 있으나, 평균 신용도는 악화 되는 경향이 있다. 
#일을 오래하는 것보다 교육 수준과 직급이 신용도에 영향을 준다. 짤리지 않고 오래 일을 할수 있다고 해서 신용도가 좋은 것은 아니다. (즉 안전성이 좋은 신용도를 보장하지 않는다.)
```
